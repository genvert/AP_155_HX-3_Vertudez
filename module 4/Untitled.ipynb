```python
import numpy as np
import matplotlib.pyplot as plt

def f(x,t):
    return -x**3 + np.sin(t)

a = 0
b = 10
N = 1000
h = (b-a)/N
x = 0

tpoints = np.arange(a,b,h)
xpoints = []

for t in tpoints:
    xpoints.append(x)
    x += h*f(x,t)
    print(x)
    
plt.plot(tpoints,xpoints)
plt.xlabel('t')
plt.ylabel('x(t)')
plt.show()
```
